```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
from sklearn.model_selection import train_test_split
```

```python
#Lectura
data = pd.read_csv("datasets\loan_approval_dataset.csv")
```

```python
print(data.columns)
```

```python
#Seleccionamos las columnas que nos interesan y cambiamos la columna del target en 0 y 1
columnas_interes= [' cibil_score', ' income_annum', ' loan_amount', ' loan_status']
datanew = data[columnas_interes]
datanew[' loan_status'] = data[' loan_status'].replace({' Approved': 0, ' Rejected': 1})
print(datanew.shape)
datanew.head(20)
```

```python
#Conteo de categorías para ver si los datos estan balanceados
print(datanew[' loan_status'].value_counts(normalize = True))
```

```python
#Ahora vamos a comprobar el maximo y el minimo de cada columna, se puede hacer de dos formas
for col in datanew.columns[0:3]:
    minimo = datanew[col].min()
    maximo = datanew[col].max()
    print(f'Columna: {col}, Minimo: {minimo}, Maximo: {maximo}')
    
#Con data describe podemos obtener tambien esa informacion
datanew.describe()    
    
```

```python
#Graficamos los datos en 3 dimensiones
pio.renderers.default = 'browser'
fig = px.scatter_3d(datanew, x=' cibil_score', y=' income_annum', z=' loan_amount', color=' loan_status', size_max=6, opacity= 0.5)
fig.show()
```

```python
#Creamos una función para extraer los datos y meterlos en un arreglo numpy
def crear_in_out(datos):
    X = datos.iloc[:,0:3].values
    Y = datos.iloc[:,3].values
    return X, Y

X, Y = crear_in_out(datanew)
print(X,Y)
```

```python
#Dividimos los datos en conjuntos de entrenamiento 70% validacion 15% y prueba 15%
def crear_train_val_test(X,Y):
    x_train, x_resto, y_train, y_resto = train_test_split (X, Y, test_size= 0.3, random_state=22)
    x_vl, x_test, y_vl, y_test = train_test_split(x_resto, y_resto, test_size=0.5, random_state=22)
    return x_train, y_train, x_vl, x_test, y_vl, y_test

#Ejecutamos la función
x_train, y_train, x_vl, x_test, y_vl, y_test = crear_train_val_test(X,Y)

#Verificar
print(x_train.shape, y_train.shape)
print(x_vl.shape, y_vl.shape)
print(x_test.shape, y_test.shape)
```

```python
#Vamos a hacer escalmiento de los datos entre 0 y 1(normalización)
from sklearn.preprocessing import MinMaxScaler
def escalar_datos(xtr, xvl, xts):
    scaler = MinMaxScaler()
    xtr_s = scaler.fit_transform(xtr)
    xvl_s = scaler.transform(xvl)
    xts_s = scaler.transform(xts)
    return xtr_s, xvl_s, xts_s

#Vamos a ejecutar la función vara obtener los datos escalados
x_train_s, x_vl_s, x_test_s = escalar_datos(x_train, x_vl, x_test)

#Verificamos los datos
print(x_train_s.min(axis=0), x_train_s.max(axis=0))
print(x_vl_s.min(axis=0), x_vl_s.max(axis=0))
print(x_test_s.min(axis=0), x_test_s.max(axis=0))
```

```python
#Vamos a crear una funcion para entrenar un modelo con un arblo de decisión (que utiliza un umbral para evaluar que camino coge)
# los cuales son inmunes al escalamiento y comparamos los resultados

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix

def arbol_entrenar_modelo(xtr, ytr, xvl, yvl):
    arbol = DecisionTreeClassifier(criterion= 'entropy',max_depth=10,random_state=45)
    arbol.fit(xtr,ytr)
    train_acc= arbol.score(xtr,ytr)
    val_acc= arbol.score(xvl,yvl)
    
    print(f'Exactitud entrenamiento/validación: {100*train_acc:.2f}%/{100*val_acc:.2f}%')
    return arbol  
```

```python
#sin escalamiento
print('Desempeño SIN escalamiento de los datos de entrada:')
tree = arbol_entrenar_modelo(x_train,y_train, x_vl, y_vl)

#Con escalamiento
print('-'*40)
print('Desempeño con escalamiento de los datos de entrada:')
tree_s = arbol_entrenar_modelo(x_train_s,y_train,x_vl_s,y_vl)
```

```python
#Ahora vamos a probar con knn que si es sensible al escalamiento, al calcular la distancia de los vecinos mas cercanos por la formula
#distancia=raizcuadrada de ((X1p-X1i)al cuad+(X2p-X2i)alcuad+(X3p-X3i)alcuad), aqui si no estan escalados la diferencia en las restas seria mucho mayor

from sklearn.neighbors import KNeighborsClassifier
def knn_entrenar_modelo(xtr, ytr, xvl, yvl):
    knn = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
    knn.fit(xtr,ytr)
    train_acc= knn.score(xtr,ytr)
    val_acc= knn.score(xvl,yvl)
    print(f'Exactitud entrenamiento/validación: {100*train_acc:.2f}%/{100*val_acc:.2f}%')
    return knn 
    
    
```

```python
#sin escalamiento
print('Desempeño SIN escalamiento de los datos de entrada:')
knn = knn_entrenar_modelo(x_train,y_train, x_vl, y_vl)

#Con escalamiento
print('-'*40)
print('Desempeño con escalamiento de los datos de entrada:')
knn_s = knn_entrenar_modelo(x_train_s,y_train,x_vl_s,y_vl)
```

```python
#Ahora vamos a hacerlo con una red neuronal, que tambien es sensible, porque cuanto mas grande sea el dato al multiplicar por el peso va a tener mayor valor
#Verbose 0 es para que no imprima el entrenamiento
import tensorflow as tf 

def red_neuronal_entrenar_modelo(xtr, ytr, xvl, yvl):
    tf.keras.utils.set_random_seed(45)
    
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(units=32, activation='relu', input_shape=(3,)),
        tf.keras.layers.Dense(units=16, activation='relu'),
        tf.keras.layers.Dense(units=1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='BinaryCrossentropy', metrics = ['accuracy'])
    model.fit(xtr,ytr, epochs=10, verbose=0)
    train_loss, train_acc= model.evaluate(xtr,ytr, verbose=0)
    val_loss, val_acc= model.evaluate(xvl,yvl, verbose=0)
    
    print(f'Exactitud entrenamiento/validación: {100*train_acc:.2f}%/{100*val_acc:.2f}%')
    return model
    
```

```python
#sin escalamiento
print('Desempeño SIN escalamiento de los datos de entrada:')
red_neuron = red_neuronal_entrenar_modelo(x_train,y_train, x_vl, y_vl)

#Con escalamiento
print('-'*40)
print('Desempeño con escalamiento de los datos de entrada:')
red_neuron_s = red_neuronal_entrenar_modelo(x_train_s,y_train,x_vl_s,y_vl)
```

```python
#Coeficientes Red Neuronal SIN escalamiento
red_neuron.layers[1].get_weights()
```

```python
#Prediccion con el modelo que mejor resultados a dado
y_pred = red_neuron_s.predict(x_test_s)
y_pred = (y_pred>=0.5).astype(int)
print(y_pred)
cm = confusion_matrix(y_test, y_pred)
num_one = sum(y_pred)
num_onetr = sum(y_test)
print(num_onetr)
print(num_one)
print(cm)
```
